```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
df = pd.read_csv('sample_parameters.csv')

# Set up the 1×3 grid of scatter plots
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Plot combinations
axes[0].scatter(df['YoungsModulus3'], df['Density1'], alpha=0.7)
axes[0].set_xlabel("YoungsModulus3")
axes[0].set_ylabel("Density1")
axes[0].set_title("YoungsModulus3 vs Density1")
axes[0].grid(True, linestyle="--", alpha=0.5)

axes[1].scatter(df['YoungsModulus3'], df['Density3'], alpha=0.7)
axes[1].set_xlabel("YoungsModulus3")
axes[1].set_ylabel("Density3")
axes[1].set_title("YoungsModulus3 vs Density3")
axes[1].grid(True, linestyle="--", alpha=0.5)

axes[2].scatter(df['Density1'], df['Density3'], alpha=0.7)
axes[2].set_xlabel("Density1")
axes[2].set_ylabel("Density3")
axes[2].set_title("Density1 vs Density3")
axes[2].grid(True, linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Halton samples without header
df = pd.read_csv('halton_15d_150.csv', header=None)
df.columns = [f'x{i+1}' for i in range(df.shape[1])]

# Select three dimensions
cols = ['x1', 'x2', 'x3']
data = df[cols]

# 2D scatter 1: x1 vs x2
plt.figure()
plt.scatter(data['x1'], data['x2'], s=15)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Halton Sample: x1 vs x2')
plt.tight_layout()
plt.show()

# 2D scatter 2: x1 vs x3
plt.figure()
plt.scatter(data['x1'], data['x3'], s=15)
plt.xlabel('x1')
plt.ylabel('x3')
plt.title('Halton Sample: x1 vs x3')
plt.tight_layout()
plt.show()

# 2D scatter 3: x2 vs x3
plt.figure()
plt.scatter(data['x2'], data['x3'], s=15)
plt.xlabel('x2')
plt.ylabel('x3')
plt.title('Halton Sample: x2 vs x3')
plt.tight_layout()
plt.show()

```
